```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
from datasets import load_dataset
#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot
```

```python
df_US_pivot = df_US_pivot.reset_index()
df_US_pivot['month_number'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot
```

```python
df_US_pivot = df_US_pivot.sort_values(by='month_number')
df_US_pivot = df_US_pivot.set_index('job_posted_month')
df_US_pivot = df_US_pivot.drop(columns='month_number')
df_US_pivot
```

```python
top_4 = df['job_title_short'].value_counts().head(4)
top_4 = top_4.index.tolist()
top_4
```

```python
df_US_pivot[top_4].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.ylabel('Job Count')
plt.xlabel('2023')
```

```python
#THE FOLLOWING IS AN INSTANCE WHERE TWO DIFFERENT DATAFRAMES WERE MERGED USING .merge()

# Data 2 source

df_US_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
df_US_software_pivot
```

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
df_US_merged
```

```python
top_5_merged = (df_US_merged.sum()
                .sort_values(ascending=False)
                .head()
                .index
                .to_list()
            )
top_5_merged
```

```python
df_US_merged[top_5_merged].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Jobs in the US')
plt.ylabel('Job Count')
plt.xlabel('2023')
plt.ylim(2000, 20000)
plt.legend()
```
