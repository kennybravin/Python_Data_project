```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
from datasets import load_dataset
#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_title_short'].value_counts()
```

```python
df.info()
```

```python
meadian_yearly_salary = df['salary_year_avg'].median()
median_hourly_salary = int(round(df['salary_hour_avg'].median()))
```

```python
df_filled = df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(meadian_yearly_salary)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_hourly_salary)
```

```python
df_unique = df_filled
df_unique = df_filled.drop_duplicates(subset=['job_title', 'job_location', 'company_name'])
```
