```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
from datasets import load_dataset
#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted month'] = df['job_posted_date'].dt.month
monthly_counts = df['job_posted month'].value_counts().sort_index()
```

```python
plt.plot(monthly_counts.index, monthly_counts)
```

```python
job_counts = df['job_title_short'].value_counts()
plt.title('Postings by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45, ha='right')
plt.bar(job_counts.index, job_counts)
```

```python
job_counts = job_counts.sort_values(ascending=True)
plt.barh(job_counts.index, job_counts)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

job_counts = job_counts.sort_values(ascending=True)

plt.figure(figsize=(10,6))
plt.barh(job_counts.index, job_counts)

# Add comma separator to x-axis values
plt.gca().xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.title('Pstings by Job Title')
plt.xlabel('Count of Job Postings')
plt.ylabel('Job Title')

plt.show()
```
