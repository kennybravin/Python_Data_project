```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
from datasets import load_dataset
#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

job_counts = df['job_title_short'].value_counts()
```

```python
job_counts.plot(kind='bar')
plt.title('Postings by Job Title')
plt.xlabel(' ')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45, ha='right')
```

```python
df[['job_posted_date', 'salary_year_avg']].dropna(subset=['salary_year_avg'])
```

```python
median_salaries_group = df.groupby('job_title_short')[['salary_year_avg']].median().sort_values('salary_year_avg')
```

```python
median_salaries_group.plot(kind='barh')
plt.title('Median Salary by Job Title')
plt.xlabel('Salary ($USD)')
plt.ylabel('')
```
